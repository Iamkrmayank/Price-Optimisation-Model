```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/Users/kumarmayank/Downloads/Dataset_modified_pom.csv')
df = df.drop(['Product Availibility index','City'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Naive forecasting
df['MRP_Pred_Naive'] = df['MRP'].shift(1)
df['SP_Pred_Naive'] = df['SP'].shift(1)
df['MSP_Pred_Naive'] = df['MSP'].shift(1)

# Seasonal Naive forecasting (assuming a monthly seasonality)
df['MRP_Pred_SeasonalNaive'] = df['MRP'].shift(12)
df['SP_Pred_SeasonalNaive'] = df['SP'].shift(12)
df['MSP_Pred_SeasonalNaive'] = df['MSP'].shift(12)

# Create the time series plot
plt.figure(figsize=(12, 8))

# Plotting the actual data
plt.plot(df['MRP'], label='MRP (Actual)', color='red')
plt.plot(df['SP'], label='SP (Actual)', color='green')
plt.plot(df['MSP'], label='MSP (Actual)', color='blue')

# Plotting Naive predictions
plt.plot(df['MRP_Pred_Naive'], label='MRP Prediction (Naive)', linestyle='dashed', color='orange')
plt.plot(df['SP_Pred_Naive'], label='SP Prediction (Naive)', linestyle='dashed', color='purple')
plt.plot(df['MSP_Pred_Naive'], label='MSP Prediction (Naive)', linestyle='dashed', color='brown')

# Plotting Seasonal Naive predictions
plt.plot(df['MRP_Pred_SeasonalNaive'], label='MRP Prediction (Seasonal Naive)', linestyle='dashed', color='cyan')
plt.plot(df['SP_Pred_SeasonalNaive'], label='SP Prediction (Seasonal Naive)', linestyle='dashed', color='magenta')
plt.plot(df['MSP_Pred_SeasonalNaive'], label='MSP Prediction (Seasonal Naive)', linestyle='dashed', color='yellow')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Plotting of MRP, SP, and MSP with Naive and Seasonal Naive Predictions')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Show the plot
plt.show()
```
