```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is your DataFrame with columns "DR1", "DR2", "DR3", "MRP", "SP", "MSP"

# Iterate through rows and plot scatter with linear regression line
for index, row in df.iterrows():
    # Extract data points
    x_values = row[["MRP", "SP", "MSP"]]
    y_values = row[["DR1", "DR2", "DR3"]]

    # Check for missing values
    if x_values.isnull().any() or y_values.isnull().any():
        print(f"Skipping row {index} due to missing values.")
        continue

    # Check for numeric values
    if not np.issubdtype(x_values.dtype, np.number) or not np.issubdtype(y_values.dtype, np.number):
        print(f"Skipping row {index} due to non-numeric values.")
        continue

    # Fit linear regression
    coefficients = np.polyfit(x_values, y_values, 1)
    poly = np.poly1d(coefficients)

    # Plot scatter and linear regression line
    plt.scatter(x_values, y_values, label=f'Row {index}')
    plt.plot(x_values, poly(x_values), label=f'Regression Line {index}')

# Add labels and legend
plt.xlabel('Price')
plt.ylabel('DRs')
plt.legend()
plt.title('Scatter plot with Linear Regression Lines')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Feature Engineering
df['day_of_week'] = df['Date'].dt.dayofweek
df['month'] = df['Date'].dt.month
df['quarter'] = df['Date'].dt.quarter
```

```python
# Train-Test Split
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

# ARIMA Model
model = ARIMA(train['MRP'], order=(5,1,0))
fit_model = model.fit()
```

```python
# Predictions
predictions = fit_model.predict(start=len(train), end=len(train)+len(test)-1, typ='levels')

# Model Evaluation
rmse = sqrt(mean_squared_error(test['MRP'], predictions))
print(f'Root Mean Squared Error (RMSE): {rmse}')

# Visualization
plt.plot(train['MRP'], label='Train')
plt.plot(test['MRP'], label='Test')
plt.plot(test.index, predictions, label='Predictions', linestyle='dashed')
plt.legend()
plt.savefig('time-series.png')
plt.show()
```
