```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel("Dataset_modified_pom.xlsx")
```

```python
df = df.drop(['Product Availibility index','City'],axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
df['Median'] = df[['MRP', 'SP', 'MSP']].median(axis=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming df is your original DataFrame with 'Date', 'MRP', 'SP', and 'MSP' columns

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Downsample by aggregating into monthly averages
df_resampled = df.resample('M', on='Date').mean()

# Calculate the median for each time series for each date
median_values = df_resampled.groupby(df_resampled.index)['MRP', 'SP', 'MSP'].median()

# Create subplots with 2 rows and 1 column
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12), sharex=True)

# Plotting the time series data on the first subplot
ax1.plot(df_resampled['MRP'], label='MRP', color='red')
ax1.plot(df_resampled['SP'], label='SP', color='green')
ax1.plot(df_resampled['MSP'], label='MSP', color='blue')
ax1.set_ylabel('Price')
ax1.set_title('Time Series Plot of MRP, SP, and MSP')
ax1.legend()

# Scatter plot of Median vs Date on the second subplot
ax2.scatter(median_values.index, median_values['MRP'], label='Median MRP', color='red')
ax2.scatter(median_values.index, median_values['SP'], label='Median SP', color='green')
ax2.scatter(median_values.index, median_values['MSP'], label='Median MSP', color='blue')
ax2.set_xlabel('Date')
ax2.set_ylabel('Median Price')
ax2.set_title('Scatter Plot of Median MRP, SP, and MSP')
ax2.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming df is your original DataFrame with 'Date', 'MRP', 'SP', and 'MSP' columns

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Downsample by aggregating into monthly averages
df_resampled = df.resample('M', on='Date').mean()

# Calculate the mean for each time series for each date
mean_values = df_resampled.groupby(df_resampled.index)['MRP', 'SP', 'MSP'].mean()

# Create subplots with 2 rows and 1 column
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 12), sharex=True)

# Plotting the time series data on the first subplot
ax1.plot(df_resampled['MRP'], label='MRP', color='red')
ax1.plot(df_resampled['SP'], label='SP', color='green')
ax1.plot(df_resampled['MSP'], label='MSP', color='blue')
ax1.set_ylabel('Price')
ax1.set_title('Time Series Plot of MRP, SP, and MSP')
ax1.legend()

# Scatter plot of Mean vs Date on the second subplot
ax2.scatter(mean_values.index, mean_values['MRP'], label='Mean MRP', color='red')
ax2.scatter(mean_values.index, mean_values['SP'], label='Mean SP', color='green')
ax2.scatter(mean_values.index, mean_values['MSP'], label='Mean MSP', color='blue')
ax2.set_xlabel('Date')
ax2.set_ylabel('Mean Price')
ax2.set_title('Scatter Plot of Mean MRP, SP, and MSP')
ax2.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
